```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  
y = np.array([0, 0, 0, 1])                     
```

```python
clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=45)
clf_tree.fit(X, y)
```

```python
from sklearn.linear_model import Perceptron
```

```python
clf_perceptron = Perceptron()
clf_perceptron.fit(X, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_decision_boundary(model, X, y, ax, title):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    ax.set_title(title)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

plot_decision_boundary(clf_tree, X, y, ax[0], 'Decision Tree (Entropy, max_depth=2) Decision Boundary')

plot_decision_boundary(clf_perceptron, X, y, ax[1], 'Perceptron Decision Boundary')

plt.tight_layout()
plt.show()


tree_preds = clf_tree.predict(X)
perceptron_preds = clf_perceptron.predict(X)

print("Decision Tree Predictions:", tree_preds)
print("Perceptron Predictions:", perceptron_preds)
```
